```python
# 표로 만들기 위해서 사용한다.
# 그래프를 위해서 만드는 라이브러리는 matplotlib, 표를 만드는 라이브러리는 pandas
# NumPy array는 한가지 데이터 타입만 표현이 가능해서 여러가지를 보여줄 수 없다. -> pandas
# pandas: 데이터를 불러오기, 데이터를 가공, 데이터 분석, 데이터 시각화
import pandas as pd

df = pd.DataFrame({
  # key, value로 이루어져있다. array 형태로
    'category': ['skirt', 'sweater', 'coat', 'jeans'],
    'quantity': [10, 15, 6, 11],                                     
    'price': [30000, 60000, 95000, 35000]
})

df

df['category']

# DataFrame은 여러개의 Series가 조합된것
type(df['category'])

df['quantity'].mean()
df['quantity'] * df['price']
```
